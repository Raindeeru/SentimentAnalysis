```python
import pandas as pd #for adding dataframe support
from sklearn.model_selection import train_test_split #for splitting the data set into training and testing data
from sklearn.feature_extraction.text import TfidfVectorizer #for vectorization of dataset
from sklearn.linear_model import LogisticRegression #prediction model used for this dataset which contains categorical data
from sklearn.metrics import accuracy_score, classification_report #for measuring statistics for the data
import nltk
from nltk.corpus import stopwords
import string
from textblob import TextBlob
from textblob import Word
from nltk.tokenize import word_tokenize 

# Download necessary NLTK resources (run this once)
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('punkt_tab')
```

```python
drugs = pd.read_csv('drugLibTrain_raw.tsv', sep = '\t') #pd.read_csv is used to load the dataframe however since the datta is in .tsv format, sep= '\t'
                                                        #is used because the values of a .tsv file is separated by tabs
```

```python
drugs.info()
```

```python
clean_drugs = drugs.dropna(subset=['sideEffectsReview'])

clean_drugs.info()
```

```python
def splitstring(x):
    x = " ".join(x.lower() for x in x.split()) 
    return x

clean_drugs['sideEffectsReview'] = clean_drugs['sideEffectsReview'].apply(splitstring)

clean_drugs.head(5)
```

```python
clean_drugs['sideEffectsReview'] = clean_drugs['sideEffectsReview'].str.replace('[^\w\s]', "", regex = True)

clean_drugs.head(5)
```

```python
stop = stopwords.words('english')
def replace_stopwords(x):
    x = " ".join(x for x in x.split() if x not in stop)
    return x

    
clean_drugs['sideEffectsReview'] = clean_drugs['sideEffectsReview'].apply(replace_stopwords)
```

```python
clean_drugs
```

```python
def lemmatizedrugs(x):
    x = " ".join([Word(word).lemmatize() for word in x.split()])
    return x

clean_drugs['sideEffectsReview'] = clean_drugs['sideEffectsReview'].apply(lemmatizedrugs)

```

```python
clean_drugs
```

```python
def tokenizedrugs(x):
    return word_tokenize(x) if isinstance(x,str) else[]
    
clean_drugs['sideEffectsReview'] = clean_drugs['sideEffectsReview'].apply(tokenizedrugs)
```

```python
clean_drugs['sideEffectsReview'].head(10)
```

```python
X = clean_drugs['sideEffectsReview']
y = clean_drugs['sideEffects']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size = 0.2, random_state = 20, stratify = y
)
```

```python
X_train.head(10)
```

```python
X_train = [' '.join(tokens) if isinstance(tokens, list) else tokens for tokens in X_train]
X_test = [' '.join(tokens) if isinstance(tokens, list) else tokens for tokens in X_test]
```

```python

from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(
    lowercase=True, stop_words='english', max_df=0.95, min_df=2, ngram_range=(1, 2))

X_train_tfidf = vectorizer.fit_transform(X_train)

X_test_tfidf = vectorizer.transform(X_test)

```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score

# Assuming you have your training data (X_train_tfidf, y_train) and testing data (X_test_tfidf, y_test) ready

# Choose and Instantiate the Model:
model = LogisticRegression(random_state=42, solver='liblinear') # You can experiment with different solvers

# Train the Model:
model.fit(X_train_tfidf, y_train)

# Make Predictions:
y_pred = model.predict(X_test_tfidf)

# Evaluate the Model (optional but highly recommended):
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy:.4f}")
print("Classification Report:\n", report)
```
